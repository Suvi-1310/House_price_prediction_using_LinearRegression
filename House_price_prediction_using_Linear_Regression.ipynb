###load_the_dataset

```python
house_price_prediction = pd.read_csv('house prediction.csv')
house_price_prediction.head()
```

###plot_the_graph_between_area_and_price

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(house_price_prediction.Area,house_price_prediction.Price,color = 'blue', marker = '*')
```

###Segregate_the_dataset_into_x_and_y

```python
x = house_price_prediction.drop('Price', axis = 'columns')
x
```

```python
y = house_price_prediction.Price
y
```

###training-LinearRegression(single_variable)

```python
model = LinearRegression()
model.fit(x,y)
```

###Prediction_of_the_model

```python
x = 6953
LandAreainSqFt = [[x]]
PredictedModelResult = model.predict(LandAreainSqFt)
PredictedModelResult
```

###algorithm-LinearRegression(single_variable)

######straight_line - (y = m * x + c)
######where m = coefficient and c = intercept

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
y = m * x + c
print("The price of {0} square feet land is: {1}".format(x,y[0]))
```
